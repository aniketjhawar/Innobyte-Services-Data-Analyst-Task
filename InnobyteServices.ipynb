```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Amazon Sale Report.csv')
print(df.head())
```

```python
total_sales = df['Amount'].sum()
print(f"Total Sales: {total_sales}")

```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.resample('M')['Amount'].sum().plot(kind='line')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.show()
```

```python
product_categories = df['Category'].value_counts()
print(product_categories)

```

```python
product_categories.plot(kind='bar')
plt.title('Product Categories Distribution')
plt.xlabel('Product Category')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()

```

```python
fulfillment_methods = df['Fulfilment'].value_counts()
print(fulfillment_methods)

```
